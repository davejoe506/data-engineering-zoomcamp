###  Question 3

The Python code below ingests the given data into Postgres. My solution (8,007) contains SQL, so please see README.

```python
import os
import pandas as pd
from sqlalchemy import create_engine
import psycopg2

connection_string = 'postgresql://root:root@localhost:5432/ny_taxi'
engine = create_engine(connection_string)

# trip data configuration
trips_url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet'
trips_file = 'green_tripdata_2025-11.parquet'
trips_table = 'green_taxi_trips'

# zone data configuration
zones_url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv'
zones_file = 'taxi_zone_lookup.csv'
zones_table = 'taxi_zone_lookup'
```

```python
# download trip data
if not os.path.exists(trips_file):
    print(f"Downloading {trips_url}...")
    import urllib.request
    urllib.request.urlretrieve(trips_url, trips_file)
else:
    print(f"Trip data file '{trips_file}' already exists.")

# download zone data
if not os.path.exists(zones_file):
    print(f"Downloading {zones_url}...")
    import urllib.request
    urllib.request.urlretrieve(zones_url, zones_file)
else:
    print(f"Zone data file '{zones_file}' already exists.")
```

```python
# load trip data
print(f"Loading trips from '{trips_file}'...")
df_trips = pd.read_parquet(trips_file)

# convert datetimes to ensure correct handling
df_trips['lpep_pickup_datetime'] = pd.to_datetime(df_trips['lpep_pickup_datetime'])
df_trips['lpep_dropoff_datetime'] = pd.to_datetime(df_trips['lpep_dropoff_datetime'])
print(f"Trips loaded. Shape: {df_trips.shape}")

# load zone data
print(f"Loading zones from '{zones_file}'...")
df_zones = pd.read_csv(zones_file)
print(f"Zones loaded. Shape: {df_zones.shape}")
```

```python
# ingest trip data
print(f"Ingesting '{trips_table}'...")
df_trips.head(0).to_sql(name=trips_table, con=engine, if_exists='replace')
df_trips.to_sql(name=trips_table, con=engine, if_exists='append', chunksize=100000)
print("Trip data ingestion complete.")

# ingest zone data
print(f"Ingesting '{zones_table}'...")
df_zones.to_sql(name=zones_table, con=engine, if_exists='replace', index=False)
print("Zone data ingestion complete.")
```
